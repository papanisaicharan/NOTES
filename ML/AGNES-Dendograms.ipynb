```python
import numpy as np

X = np.array([[5,3],  
    [10,15],
    [15,12],
    [24,10],
    [30,30],
    [85,70],
    [71,80],
    [60,78],
    [70,55],
    [80,91],])
```

```python
import matplotlib.pyplot as plt

labels = range(1, 11)  
plt.figure(figsize=(10, 7))  
plt.subplots_adjust(bottom=0.1)  
plt.scatter(X[:,0],X[:,1], label='True Position')

for label, x, y in zip(labels, X[:, 0], X[:, 1]):  
    plt.annotate(
        label,
        xy=(x, y), xytext=(-3, 3),
        textcoords='offset points', ha='right', va='bottom')
plt.show() 
```

let's draw the dendrograms for our data points. We will use the scipy library for that purpose. Execute the following script:

```python
from scipy.cluster.hierarchy import dendrogram, linkage  
from matplotlib import pyplot as plt
```

### scipy.cluster.hierarchy

The hierarchy module provides functions for hierarchical and agglomerative clustering. Its features include generating hierarchical clusters from distance matrices, calculating statistics on clusters, cutting linkages to generate flat clusters, and visualizing clusters with dendrograms.

https://docs.scipy.org/doc/scipy/reference/cluster.hierarchy.html#module-scipy.cluster.hierarchy

linkage(y[, method, metric, optimal_ordering])	Perform hierarchical/agglomerative clustering.

try to read this: https://nlp.stanford.edu/IR-book/html/htmledition/contents-1.html

Flat clustering creates a flat set of clusters without any explicit structure that would relate clusters to each other. Hierarchical clustering creates a hierarchy of clusters

dendrogram(Z[, p, truncate_mode, …])	Plot the hierarchical clustering as a dendrogram

```python
linked = linkage(X, 'single')
# there will be three types of linkage
# single  : min distance is considered for clustering
# average : avg distance is considered for clustering
# complete: max distance is considered for clustering

labelList = range(1, 11)

plt.figure(figsize=(10, 7))  
dendrogram(linked,  
            orientation='top',
            labels=labelList,
           )
plt.show() 
```

```python
# know coming to assignment
# https://docs.scipy.org/doc/scipy/reference/generated/scipy.cluster.hierarchy.dendrogram.html#scipy.cluster.hierarchy.dendrogram
#
import numpy as np

from scipy.cluster.hierarchy import dendrogram, linkage
from scipy.spatial.distance import squareform

import matplotlib.pyplot as plt


mat = np.array([[0,9,3,6,11],[9,0,7,5,10],[3,7,0,9,2],[6,5,9,0,8],[11,10,2,8,0]])
dists = squareform(mat)
# https://docs.scipy.org/doc/scipy/reference/spatial.distance.html
# squareform(X[, force, checks])	Convert a vector-form distance vector to a square-form distance matrix, and vice-versa.
linkage_matrix = linkage(dists, "complete")
# there will be three types of linkage
# single  : min distance is considered for clustering
# average : avg distance is considered for clustering
# complete: max distance is considered for clustering, ward is also present
# https://people.revoledu.com/kardi/tutorial/Clustering/Linkages.htm

dendrogram(linkage_matrix, labels=[ "1", "2", "3","4","5"])
plt.title("complete linkage")
plt.show()
```
